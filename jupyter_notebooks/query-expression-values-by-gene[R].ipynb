# Using Genestack Omics APIs for querying expression values

### Set instance and token

```r
library(integrationCurator)
library(expressionUser)
library(preparationUser)
library(stringi)
suppressMessages(library(tidyverse))
suppressMessages(library(integrationCurator)) # Genestack client library

Sys.setenv(PRED_SPOT_HOST = 'br-dev.genestack.net',
           PRED_SPOT_TOKEN = 'tknRoot',
           PRED_SPOT_VERSION = 'default-released')
```

### Get samples by study accession

```r
system.time(
    res <- OmicsQueriesApi_search_expression_data(sample_filter='Sex=F', ex_query='Gene=CD3E')$content$data
)
res <- split(res, res$groupId)
index = 1
info = res[[index]]
tbl = list()
tbl$runId <- info$runId
tbl$expression <- info$expression
tbl$preparations <- info$relations$preparation
tbl <- as_tibble(tbl)
info$groupId[[1]]
tbl
```

```r
# /preparations/{id}
preparation_id = tbl$preparations[[1]]
preparation_id
preparation <- PreparationSPoTApi_get_preparation(id=preparation_id)$content
preparation_group_id <- preparation$groupId
preparation_group_id
```

```r
# integration/link/preparation/preparations-to-samples/by/group/
system.time(preparation_to_samples <- PreparationIntegrationApi_get_preparation_links_to_samples(id=preparation_group_id)$content$data)
preparation_to_samples
```

```r
# integration/link/study/by/sample/
sample_acc <- preparation_to_samples$sampleIds[[1]][[1]]
sample_acc
system.time(study <- StudyIntegrationApi_get_study_by_sample(id=sample_acc)$content)
study
```

```r
expressions = left_join(tbl, preparation_to_samples, by=c("preparations"="preparationId"))
expressions %>% select(sampleIds, expression)
```
